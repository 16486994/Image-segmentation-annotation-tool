```python
def Leftcallback(event): 
    global img,imgpath,canvas,master,w_box,h_box,E1,v,h,L2,ft1
    if len(v)<6:
        img=Manualcut(round(event.x/2))
        Exhibition()
def Rightcallback(event): 
    global img,imgpath,canvas,master,w_box,h_box,E1,v,h,L2,ft1
    img=Manualdel(round(event.x/2))
    Exhibition()
def get_pic_path():
    files=[]
    path=os.path.abspath("./SimpleImg/")
    for _, _,name in os.walk(path):
        files.append(name)
    return files[0]
def PrevCallBack():
    global img,imgpath,canvas,master,w_box,h_box,E1,v,h,L2,ft1,pil_image,Pre_write,pre_filename
    a=[str(item[0]) for item in pre_filename]
    srcFile = os.path.abspath("./SimpleImg")+"/"+E1.get()+".png"
    dstFile = os.path.abspath("./SimpleImg")+"/"+E1.get()+ '-'+ '_'.join(a)+".png"
    pil_image.save(srcFile)
    try:
        os.rename(srcFile,dstFile)
    except Exception as e:
        print(e)
        print('rename file fail\r\n')
    else:
        print('rename file success\r\n')
    YesCallBack()    
    if (str(int(E1.get())-1)+".png") in get_pic_path():
        E1tmp=int(E1.get())-1
        E1.delete('0', END)
        E1.insert(0, E1tmp)
        imgpath=os.path.abspath("./SimpleImg")+"/"+E1.get()+".png"
        img=cut()
        Exhibition()
    else:
        messagebox.showinfo(title='Info', message='no find file')
def NextCallBack():
    global img,imgpath,canvas,master,w_box,h_box,E1,v,h,L2,ft1,pil_image,Pre_write,pre_filename,text
    if len(v)==6:
        a=[str(item[0]) for item in pre_filename]
        srcFile = os.path.abspath("./SimpleImg")+"/"+E1.get()+".png"
        dstFile = os.path.abspath("./SimpleImg")+"/"+E1.get()+ '-'+ '_'.join(a)+".png"
        pil_image.save(srcFile)
        try:
            os.rename(srcFile,dstFile)
        except Exception as e:
            print(e)
            print('rename file fail\r\n')
        else:
            print('rename file success\r\n')
        YesCallBack()
        if (str(int(E1.get())+1)+".png") in get_pic_path():
            E1tmp=int(E1.get())+1
            E1.delete('0', END)
            E1.insert(0, E1tmp)
            imgpath=os.path.abspath("./SimpleImg")+"/"+E1.get()+".png"
            img=cut()
            Exhibition()
        else:
            messagebox.showinfo(title='Info', message='no find file')
def YesCallBack():
    global img,imgpath,canvas,master,w_box,h_box,E1,v,h,L2,ft1,pil_image,Pre_write
    if (E1.get()+".png") in get_pic_path():
        imgpath=os.path.abspath("./SimpleImg")+"/"+E1.get()+".png"
        Pre_write=pil_image
        img=cut()
        Exhibition()
def Layout():
    global img,imgpath,canvas,master,w_box,h_box,E1,v,h,L2,ft1,text
    ft = tf.Font(family='Microsoft YaHei UI',size=43,weight=tf.BOLD) 
    ft1 = tf.Font(family='Microsoft YaHei UI',size=20,weight=tf.BOLD)
    L1 = Label(master, text="serch",font = ft1)
    L1.grid(row=0,column=0)
    text.set(len(v))
    L2 = Label(master, textvariable=text,font = ft1)
    L2.grid(row=2,column=1,columnspan=1)
    E1 = Entry(master, bd =1,font = ft1)
    E1.grid(row=0,column=1)
    path = Button(master, text ="Yes",command=lambda:YesCallBack(),font = ft1,compound='center')
    path.grid(row=0,column=2)
    canvas = Canvas(master,width =w_box,height = h_box,bg = 'white')
    canvas.bind("<Button-1>",Leftcallback)
    canvas.bind("<Button-3>",Rightcallback)
    canvas.grid(row=1,column=0,columnspan=3)
    Prev = Button(master, text ="Prev",command=PrevCallBack,font = ft,compound='center',width = round(master.winfo_width()/2))
    Next = Button(master, text ="Next",command=NextCallBack,font = ft,compound='center',width = round(master.winfo_width()/2))
    Prev.grid(row=2,column=0,columnspan=2,sticky=W)
    Next.grid(row=2,column=1,columnspan=2,sticky=E)
    return canvas
def Exhibition():
    global img,imgpath,canvas,master,w_box,h_box,E1,v,h,L2,ft1,text
    text.set(len(v))
    canvas.create_image(0,0,anchor=NW,image=img)
def LineFilter(v):
    op = [element for sublist in [[v[index],v[index+1]] for index in range(0,len(v)) \
    if index % 2==0 and abs((numpy.array(list(v[index]))-numpy.array(list(v[index+1]))))[0]<5] for element in sublist]
    v = list(set(v).difference(set(op)))
    v.sort()
    op = [element for sublist in [[v[index+1],v[index+2]] for index in range(0,len(v)) if index % 2==0 and \
                              abs((numpy.array(list(v[index]))-numpy.array(list(v[index+1]))))[0]>6 and \
                              abs((numpy.array(list(v[index]))-numpy.array(list(v[index+1]))))[0]<10 and index+2<=len(v)-1] for element in sublist]
    v = list(set(v).difference(set(op)))
    v.sort()
    return v
def cut():
    global img,imgpath,canvas,master,w_box,h_box,E1,v,h,L2,ft1,tmppath,pil_image,pre_filename
    v=Vercal(imgpath)
    v=LineFilter(v)
    v=mergeline(v)
    v=[tuple(nromline(list(x))) for x in v]
    pil_image =putVercalline(v,imgpath)
    pre_filename=v
    w, h = pil_image.size
    pil_image_resized = resize(w, h, w_box, h_box, pil_image)
    img = ImageTk.PhotoImage(pil_image_resized)
    return img
def Manualdel(clickX=0,draw=True):
    global img,imgpath,canvas,master,w_box,h_box,E1,v,h,L2,ft1,tmppath,pil_image,Pre_write,pre_filename
    a=[clickX,0,clickX,h]
    o=[list(abs(numpy.array(a)-numpy.array(list(item)))) for item in v]
    if len(o)!=0:
        del v[o.index(min(o))]
    pil_image = putVercalline(v,imgpath)
    pre_filename=v
    Pre_write=pil_image
    w, h = pil_image.size
    pil_image_resized = resize(w, h, w_box, h_box, pil_image)
    img = ImageTk.PhotoImage(pil_image_resized)
    return img
def Manualcut(clickX=0,draw=True):
    global img,imgpath,canvas,master,w_box,h_box,E1,v,h,L2,ft1,tmppath,pil_image,pre_filename
    a=[clickX,0,clickX,h]
    a=nromline(a)
    v.append(tuple(a))
    v.sort()
    pil_image = putVercalline(v,imgpath)
    pre_filename=v
    w, h = pil_image.size
    pil_image_resized = resize(w, h, w_box, h_box, pil_image)
    img = ImageTk.PhotoImage(pil_image_resized)
    return img
def nromline(a):
    global img,imgpath,canvas,master,w_box,h_box,E1,v,h,L2,ft1,tmppath,pil_image,pre_filename
    arr = [ x for x in range(0,200,5) ] 
    if a[0] not in arr:
        arr.append(a[0])
        arr.sort()
        aa=[arr[arr.index(a[0])-1],arr[arr.index(a[0])+1]]
        bb=[abs(arr[arr.index(a[0])-1]-a[0]),abs(arr[arr.index(a[0])+1]-a[0])]
        return [aa[bb.index(min(bb))],0,aa[bb.index(min(bb))],h]
    else:
        return a
def mergeline(a):
    op = [element for sublist in [[a[index]] for index in range(0,len(a)-1) \
    if index % 2!=0 and abs((numpy.array(list(a[index]))-numpy.array(list(a[index+1]))))[0]<10] for element in sublist]
    a = list(set(a).difference(set(op)))
    a.sort()
    return a
import matplotlib.pyplot as plt
import cv2,imageio,numpy,os
from tkinter import *   
from tkinter import messagebox
from PIL import ImageTk,Image,ImageDraw
from lib.GetSimple import gs 
from lib.ProjectionCutting import * 
import tkinter.font as tf
master = Tk()
master.title('Position split annotation tool')
path = r'./image.png'
tmppath= os.path.abspath(r'./image_0.png')
imgpath= os.path.abspath(path)
w_box = 400 
h_box = 100 
img=None 
canvas=None 
E1=None 
v=None 
h=50 
L2=None 
ft1=None 
text = StringVar() 
pil_image=None 
Pre_write=None 
pre_filename=None 
img=cut()
canvas=Layout()
Exhibition()
mainloop()
```
